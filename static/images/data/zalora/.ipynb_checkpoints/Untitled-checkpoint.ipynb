```python
import numpy as np
import pandas as pd
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("data-zalora-hand-bag.csv", error_bad_lines = False)
```

```python
df['name'] = df['name'].map(lambda x: x.replace(" ", "_"))
df['name'] = df['name'].map(lambda x: x.replace("/", "_"))
df['brand'] = df['brand'].map(lambda x: x.replace(" ", "_"))
df['brand'] = df['brand'].map(lambda x: x.replace("&", "_"))
df['category'] = df['category'].map(lambda x: x.replace(" ", "_"))
df['category'] = df['category'].map(lambda x: x.replace("&", "_"))
```

```python
for category in df['category'].unique():
    os.mkdir(category)
```

```python
for index, rows in df.iterrows():
    name = "{:06d}".format(rows['no'])
    file_dir = os.path.join(rows['category'], name)
    os.mkdir(file_dir)
```

```python
for index, rows in df.iterrows():
    if (rows['no'] < 35206):
        list_image = rows['list_image'].split("|")[:-1]
        name = "{:06d}".format(rows['no'])
        file_dir = os.path.join(rows['category'], name)
        for img in list_image:
            command = "wget " + "-P " + file_dir + " " + str(img)
            print(command)
            os.system(command)
```

```python
from os import listdir
from os.path import isfile, join

mypath = "./Dress_copy"
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
root = "./Dress_copy"
included_files = []
for path, subdirs, files in os.walk(root):
    file_list = []
    for name in files:
        if name[-5:]!="Store":
            file_list.append(os.path.join(path, name))
    file_list.sort()
    try:
        included_files.append(file_list[0])
    except:
        pass

included_files.sort()
np.save('zalora_dress', included_files)
```
